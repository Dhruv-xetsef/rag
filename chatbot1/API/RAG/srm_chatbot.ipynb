```python
import time
from concurrent.futures import ThreadPoolExecutor
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings # Assuming this is correctly imported

# --- Setup (Same as yours) ---
# NOTE: The 'documents' variable must be loaded from your previous cell outputs.
# batch_size in your original loop refers to the chunk size for adding to FAISS, 
# here we'll use a larger chunk size (e.g., 256) for parallel processing efficiency.
TOTAL_DOCUMENTS = len(documents)
print(f"Starting parallel indexing for {TOTAL_DOCUMENTS} documents...")
start_time = time.time()

# Assuming embeddings is already defined and loaded from OllamaEmbeddings(model="nomic-embed-text")
embeddings = OllamaEmbeddings(model="nomic-embed-text")

# --- Configuration for Parallelism ---
# ⚠️ ACTION REQUIRED: Adjust these values based on your CPU cores and available RAM/VRAM.
# A good starting point is 2x to 4x the number of CPU cores for I/O-bound tasks.
MAX_WORKERS = 8  # Number of parallel threads/workers
EMBEDDING_CHUNK_SIZE = 256 # How many documents each worker will process in a batch. 
                         # Keep this large for efficiency, or small if Ollama times out.
                         
# --- Step 1: Initialize FAISS with the first small batch ---
initial_batch_size = 64 
first_batch_docs = documents[:initial_batch_size]

# Ensure the database is initialized before starting parallel additions
db = FAISS.from_documents(first_batch_docs, embeddings)
print(f"Processed initial batch (0 to {initial_batch_size}).")

# --- Step 2: Parallel Document Addition ---

# Get the remaining documents
remaining_documents = documents[initial_batch_size:]

# Function to be executed in parallel by each worker
def add_documents_to_faiss(docs_chunk, chunk_index):
    """Adds a chunk of documents to the global FAISS index."""
    try:
        # FAISS.add_documents handles the embedding generation internally for its batch
        db.add_documents(docs_chunk)
        print(f"✅ Worker {chunk_index}: Processed chunk with {len(docs_chunk)} documents.")
    except Exception as e:
        print(f"❌ Worker {chunk_index} failed to process chunk: {e}")
        
# Chunk the remaining documents for parallel processing
chunks = [remaining_documents[j:j + EMBEDDING_CHUNK_SIZE] 
          for j in range(0, len(remaining_documents), EMBEDDING_CHUNK_SIZE)]

print(f"Divided remaining {len(remaining_documents)} documents into {len(chunks)} parallel chunks.")
print(f"Starting parallel processing with {MAX_WORKERS} workers...")

# Use ThreadPoolExecutor to run the additions in parallel
with ThreadPoolExecutor(max_workers=MAX_WORKERS) as executor:
    # Submit all chunks to the executor
    futures = [executor.submit(add_documents_to_faiss, chunk, i) for i, chunk in enumerate(chunks)]
    
    # Wait for all futures (tasks) to complete
    for future in futures:
        future.result() # This will raise any exceptions that occurred in the worker threads.

end_time_indexing = time.time()
print("\n--- FAISS Indexing Complete ---")
print(f"Total FAISS Index documents: {len(db.docstore._dict)}")
print(f"Total time taken: {end_time_indexing - start_time:.2f} seconds")
print("FAISS index created")
```

```python
from langchain_core.runnables import RunnablePassthrough

# This is the replacement for 'create_retrieval_chain'
retrieval_chain = (
    {"context": retriever, "input": RunnablePassthrough()}
    | document_chain
)
```
